```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

```python
excel_df = pd.read_excel('../data/interview_test.xlsx', sheet_name='DATA',skiprows=2)
```

```python
excel_df
```

```python
excel_df = excel_df.drop(['ANA'],axis=1)
```

```python
c_list = excel_df['Prime Item Desc'].values
excel_df = excel_df.drop(['Prime Item Desc'],axis=1)
```

```python
excel_df_T = excel_df.T
excel_df_T.columns = c_list
excel_df_T.head(5)
```

```python
excel_df_T.reset_index()
```

```python
excel_df_T['Year_Week'], excel_df_T['x'] = excel_df_T.index.str.split(' ', 1).str
excel_df_T = excel_df_T.reset_index(drop=True)
excel_df_T = excel_df_T.drop(['x'],axis=1)
excel_df_T.head(5)
```

```python
excel_df_T['Date'] = [datetime.datetime.strptime(x + '-1', "%Y%W-%w") for x in excel_df_T['Year_Week']]
```

```python
excel_df_T.head(5)
```

```python
plt.plot(excel_df_T.Date, excel_df_T['PRODUCT B'])
```
